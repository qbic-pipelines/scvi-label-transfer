```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
import scanpy as sc
import scvi
```

# scANVI label transfer
This notebook describes the following process:

* train a scVI model on multiple datasets
* use the scVI model to train a scANVI model on partially labeled datasets
* use the scANVI model to perform cell type labeling
* evaluate the results

More information can be found at the scvi-tools page: https://scvi-tools.org/

### Load the datasets

Load all the datasets that have been downloaded and pre-processed as described in the `README.md`.

```python


dataset1 = sc.read_h5ad("GSE124395_labeled.h5ad")
dataset2 = sc.read_h5ad("gse158723_labeled.h5ad")
dataset3 = sc.read_h5ad("healthy_cirrhotic_human_liver_labeled.h5ad")
dataset4 = sc.read_h5ad("liver_immune_cells_human_liver_labeled.h5ad")
dataset5 = sc.read_h5ad("sc_landscape_human_liver_labeled.h5ad")

dataset1.obs_names_make_unique()
dataset2.obs_names_make_unique()
dataset3.obs_names_make_unique()
dataset4.obs_names_make_unique()
dataset5.obs_names_make_unique()
```

### Concatenate datasets and only use highly variable genes

All datasets need to be concatenated to be used with scVI.
The data is also normalized and logarithmitized. Finally the top 2000 highly variable genes are selected to be used with scVI.

```python
# concatenate datasets
adata = dataset1.concatenate(dataset2, dataset3, dataset4, dataset5)
adata.layers["counts"] = adata.X.copy()
sc.pp.normalize_total(adata, target_sum=1e4)
sc.pp.log1p(adata)
adata.raw = adata  # keep full dimension safe
sc.pp.highly_variable_genes(
    adata,
    n_top_genes=2000,
    layer="counts",
    batch_key="batch",
    subset=True
)
```

# Train a scVI model

First a scVI model is generated and trained. To do this, we have to setup the `adata` object with the `scvi.data.setup_anndata()` function as seen below. Then a model is generated with 2 layers and latent space of size 30. Finally, the scVI VAE is trained. Here we have set `max_epochs` to 70, which is sufficient on that datasets. Training for too few epochs can lead to bad results.

In the commented code, you see how to load a pre-trained model instead. After training, the model is saved using `vae.save()`.

```python
scvi.data.setup_anndata(adata, 
                        layer="counts",
                        batch_key="batch",
                        labels_key="celltype")
```

```python
vae = scvi.model.SCVI(adata, n_layers=2, n_latent=30)

# alternatively, load trained model
#vae = scvi.model.SCVI.load("liver_scVI_model", adata)
```

```python
vae.train(use_gpu=True, max_epochs=70)
```

```python
vae.save("liver_scVI_model", overwrite=True)
```

## Visualize latent representations of the scVI model

Next we extract the latent representation of the data using `vae.get_latent_representation()` and then generate a UMAP plot to visualize it. Here the data is labeled by batch and celltype. Ideally, the data should be clustered by different celltypes and not so much by batch. Of course the batch also has an effect, as these are also different technologies and individuals. But the major factor should be celltypes.

As you can see below, the celltypes from different batches cluster nicely together.

```python
# get the latent representations
adata.obsm["X_scVI"] = vae.get_latent_representation()

# cluster and umap
sc.pp.neighbors(adata, use_rep="X_scVI")
sc.tl.leiden(adata)
sc.tl.umap(adata)
```

```python
sc.pl.umap(
    adata,
    color=["batch", "celltype"],
    frameon=False
)
```

# Train a scANVI model

Next we use the trained scVI model to train a new scANVI model. You can also train a scANVI model from scratch, but using a pre-trained model is the recommended way. 

First we generate a `celltype_scanvi` field and add the celltypes from datasets 0, 1, 2 and 3, but we leave the dataset 4 unlabeled so we can later try to estimate the celltypes.

```python
adata.obs["celltype_scanvi"] = 'Unknown'
# Get the labels for datasets 0, 1, 2, 3

batch_idx = adata.obs['batch'] == "0"
adata.obs["celltype_scanvi"][batch_idx] = adata.obs.celltype[batch_idx]

batch_idx = adata.obs['batch'] == "1"
adata.obs["celltype_scanvi"][batch_idx] = adata.obs.celltype[batch_idx]

batch_idx = adata.obs['batch'] == "2"
adata.obs["celltype_scanvi"][batch_idx] = adata.obs.celltype[batch_idx]

batch_idx = adata.obs['batch'] == "3"
adata.obs["celltype_scanvi"][batch_idx] = adata.obs.celltype[batch_idx]
```

```python
adata.obs['celltype_scanvi'] = adata.obs['celltype_scanvi'].astype("str")
```

Just making sure that we have some unlabeled data now. We have 8831 cells with "Unknown" label, which is the size of dataset 4.

```python
np.unique(adata.obs["celltype_scanvi"], return_counts=True)
```

Similar to scVI, we have to setup the AnnData object and also define the key of the labels (`celltype_scanvi`).
Then we create the `scanvi_model` from the scVI model (`vae`) and define the key for the unknown cell types (here: "Unknown").

The scANVI model is then trained for 50 epochs. Here it is important to train for quite a few epochs again, otherwise the performance will be bad.

```python
scvi.data.setup_anndata(
    adata,
    layer="counts",
    batch_key="batch",
    labels_key="celltype_scanvi",
)
```

When generating the scANVI model, we have to specifiy which label corresponds to the unknown class. Here we say this is the "Unknown" label.

```python
scanvi_model = scvi.model.SCANVI.from_scvi_model(vae, "Unknown")
```

```python
scanvi_model.train(max_epochs=50)
```

### Predict unknown cell types

Once the scANVI model is trained, we can predict the unknown celltypes using the ".predict()" function. Here we save the results in the `adata.obs[C_scANVI]` field.

To evaluate the results, we extract the latent representation again and generate a UMAP plot, which we color by original celltype (`celltype_scanvi`), predicted celltype (`C_scANVI`) and batch. We can see the "Unknown" cells in the first plot. In the second plot, we can see that the cells that have been labeled "Unknown" were quite successfully labeled to the correct cell type, showing that the scANVI cell type labeling worked

```python
adata.obs["C_scANVI"] = scanvi_model.predict(adata)
adata.obsm["X_scANVI"] = scanvi_model.get_latent_representation(adata)
```

```python
sc.pp.neighbors(adata, use_rep="X_scANVI")
sc.tl.umap(adata)
```

```python
sc.pl.umap(adata, color=["celltype_scanvi", "C_scANVI", "batch"], ncols=1, frameon=False)
```

### Closer look at the numbers

Finally, besides visually inspecting the UMAPs, we can look into how many cells have been labeled correctly by scANVI. Note that labeling scRNA-seq data is difficult and often done by clustering and labeling the clusters. This can lead to false labels. Furthermore the cell types in different datasets/batches might not be entirely equivalent. It is therefore to be expected that the predicted cell types never 100% agree with the manually labeled cell types - which is not to say that they are wrong.

Looking at the heatmap, it is obvious that most cells were labeled correctly. The biggest confusion can be observed between Lymphoid and Plasma cells. I had trouble correctly labeling those manually as well, and it is likely that I made some errors, or they are sometimes just not really distinguishable.

Looking at the numbers in the confusion matrix paints a similar picture, with some immune cells "incorrectly" labeled (either by me or by scANVI). The vast majority of cells have been labeled correctly.

Note that, since scANVI is a probabilistic model, it is possible to get the proababilities for the predictions, which could be used to only label cells for which the model has a high-certainty prediction.

```python
df = adata.obs.groupby(["celltype", "C_scANVI"]).size().unstack(fill_value=0)
df = df / df.sum(axis=0)

plt.figure(figsize=(8, 8))
_ = plt.pcolor(df)
_ = plt.xticks(np.arange(0.5, len(df.columns), 1), df.columns, rotation=90)
_ = plt.yticks(np.arange(0.5, len(df.index), 1), df.index)
plt.xlabel("Predicted")
plt.ylabel("Observed")
```
